```python
import seaborn as sns


# Plot distribution plot for "founded year"
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Founded', kde=True)
plt.title('Distribution of Founded Year')
plt.xlabel('Founded Year')
plt.ylabel('Count')
plt.show()

# Plot box plot for "count of enrollment"
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='Enrollment')
plt.title('Box Plot of Count of Enrollment')
plt.ylabel('Count of Enrollment')
plt.show()
```

```python
data.corr()
```

```python
# Plot scatter plot for "rank" vs "count of enrollment"
plt.figure(figsize=(10, 6))
plt.scatter(data['Rank'], data['Enrollment'])
plt.title('Relationship between Rank and Count of Enrollment')
plt.xlabel('Rank')
plt.ylabel('Count of Enrollment')
plt.show()
```

```python
# Plot scatter plot for "rank" vs "founded year"
plt.figure(figsize=(14, 6))
plt.scatter(data['Founded'], data['Rank'])
plt.title('Relationship between Rank and Count of Enrollment')
plt.xlabel('Founded year')
plt.xticks(rotation=90)
plt.ylabel('Rank')
plt.show()
```

```python
# Plot scatter plot for "Enrollment" vs "founded year"
plt.figure(figsize=(14, 6))
plt.scatter(data['Founded'] , data['Enrollment'])
plt.title('Relationship between Rank and Count of Enrollment')
plt.xlabel('Founded year')
plt.xticks(rotation=90)
plt.ylabel('Count of Enrollment ')
plt.show()
```

## How does the count of enrollment vary between public and private universities?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Affiliation', y='Enrollment')
plt.title('Count of Enrollment by Affiliation')
plt.xlabel('Affiliation')
plt.ylabel('Count of Enrollment')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Affiliation', y='Rank')
plt.title('Count of Enrollment by Rank')
plt.xlabel('Affiliation')
plt.ylabel('Count of Rank')
plt.show()
```

## Does the continent have an influence on the rank of universities?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Continent', y='Rank')
plt.title('University Rank by Continent')
plt.xlabel('Continent')
plt.ylabel('Rank')
plt.show()
```
